```python
import numpy as np
import pandas as pd
from math import ceil
```

```python
#defini o total da população que tenho e o numero de amostras que desejo para a minha amostragem
populacao = 150
amostra   = 14

#ceil da lib math usado para arredondar os numeros
k = ceil(populacao / amostra)

#geração de um numero randomico R para a qtde de amostras que obti na variavel k
r = np.random.randint(low = 1, high = k + 1, size = 1)

#esta variavel terá o valor randomico inteiro obtido atraves do array R
#desta forma sera melhor trabalhar dentro do proximo lote de comandos
acumulador = r[0]
```

```python
sorteados = []

#para i no range de amostras que é 14
for i in range(amostra):
    sorteados.append(acumulador)
    acumulador += k
    
```

```python
#exibindo as 14 ocorrencias de amostragem sistemica 
sorteados
```

```python
#obtendo os dados do dataframe iris da amostra que foi sorteada

#carregando o dataframe
base = pd.read_csv("Dados/iris.csv")

#subset final obtendo somentes os id sorteados
base_final = base.loc[sorteados]
```
